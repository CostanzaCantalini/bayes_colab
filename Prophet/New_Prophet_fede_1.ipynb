###PROPHET

```python
from prophet import Prophet
```

```python
post=(np.zeros((49,365,1200)))
for i in range(49):
    data_prophet = df[["Date",df.columns[i+1]]]
    data_prophet.columns = ["ds","y"]
    m = Prophet(mcmc_samples=300,yearly_seasonality=True)
    m.fit(data_prophet)
    future = m.make_future_dataframe(periods=0)
    post[i,:,:] = m.predictive_samples(future)["yhat"]

```
