## Prophet Model
We model the time series $y(t)$ as the sum of 3 trends:

* $\textit{growth}$ term $g(t)$ (models non periodic changes); 
* $\textit{seasonality}$ term $s(t)$ (models periodic changes); 
* $\textit{holiday}$ term $h(t)$ (models extraordinary terms). 

\begin{equation*}
\begin{aligned}
    y(t) & = g(t) + s(t) + h(t) + \epsilon_t \;\;\;\; t = 1, ..., T \\
    \epsilon_t & \overset{\text{iid}}{\sim} \mathcal{N}(0, \sigma_{noise}^2)
\end{aligned}   
\end{equation*}

$\newline$



### Saturating (Logistic) Growth Trend
Used for forecasting problems showing a saturating growth trend:
\begin{equation*}
\begin{aligned}
    g(t) = \frac{C}{1 + exp(-k(t-m))}
\end{aligned}   
\end{equation*}
where $C$ is the maximal capacity (horizontal asymptote for $g(t)$), $k$ is the growth rate and $m$ is an offset param.

I can introduce a non constant capacity $C(t)$ (known a priori) and a changing growth rate $k(t)$.

I define $S$ changepoints at times $s_j, \; j=1,...,S$ and a vector of adjustments of the rates $\underline{\delta} \in \mathbb{R}^S$ $s.t.$ the growing rate becomes:

\begin{equation*}
    k(t) = k + \sum_{j \, : \, t>s_j}^{S} \delta_j \;\;\;\; (k \; base \; growth \; rate)
\end{equation*}

or, equivalently, 

\begin{equation*}
    k(t) = k + \underline{a(t)}^T \underline{\delta}
\end{equation*}
$$
  a(t)_j = \cases{
      1       & $ t \geq s_j$ \cr
      0       & $otherwise$ }
$$
To make the function continuous in the change points, i have to set a correction parameter $\underline{\gamma}$ (see the paper for its expression so that the model is:
\begin{equation*}
    g(t) = \frac{C(t)}{1 + exp( \, (k + \underline{a(t)}^T \underline{\delta}) \, (t - (m + \underline{a(t)}^T \underline{\gamma})) \,)}
\end{equation*}


### Linear Growth Trend

Used for forecasting problems showing a non saturating growth trend but just piece-wise constant or linear growth trends.
\begin{equation*}
    g(t) = (k + \underline{a(t)}^T \underline{\delta})t + (m + \underline{a(t)}^T \underline{\gamma})
\end{equation*}
with $\gamma_j = - s_j \delta_j$

$\newline$
In both cases, the changepoints $s_j$ can be fixed a priori if we have available information and, once they are fixed, I can make the algorithm to select them automatically their rates setting a sparse prior for \underline{\gamma} such as:
\begin{equation*}
    \delta_j \overset{\text{iid}}{\sim} \mathcal{Laplace}(0, \tau)
\end{equation*}
where parameter $\tau$ (with a prior) constrols the flexibility of the model in altering its rates. If $\tau$ goes to 0, the t reduces to standard (not-piecewise) logistic or linear growth whereas, for large tau, the model adapts too well to the data and overfits.
$\newline$

Assuming that the future will see the same average frequency and magnitude of rate changes that were seen in the history, changepoints in future events are sampled at time steps $j$ evenly spaced in time as the average frequency of changes, following the law:
\begin{equation*}
\begin{aligned}
    \delta_j & \overset{\text{iid}}{\sim} \mathcal{Laplace}(0, \lambda) \\
    \lambda & = \frac{1}{S} \sum_{j=1}^{S} \left| \delta_j \right|  \;\;\;\; (estimated \; from \; sampled \; \delta_j)
\end{aligned}
\end{equation*}

### Seasonality Trend
To model the periodicity of the trends we use the Fourier expansion of a periodic signal:

\begin{equation*}
    s(t) = \sum_{n=1}^{N} \left( a_n cos \left(\frac{2 \pi n t}{P} \right) + b_n sin \left(\frac{2 \pi n t}{P} \right) \right)
\end{equation*}

with $P$ the expected period of the phenomenon and $2N$ parameters $\underline{\beta} = \left[ a_1, b_1, ..., a_n, b_n \right]^T $.

Choosing a certain $N$ and $P$ and constructing the matrix of seasonality vector (that play the role of regressors) $\textbf{X(t)}$ for each value of $t = t_1, ..., t_T$ in the series we obtain:
\begin{equation*}
\begin{aligned}
    regressor \; vector: \;\;X(t) & = \left[ cos \left(\frac{2 \pi (1) t}{P} \right), ...,  sin \left(\frac{2 \pi (10) t}{P} \right)\right] \\
    s(t) & = X(t) \underline{\beta} \\
    prior: \;\; \underline{\beta}_i & \overset{\text{iid}}{\sim} \mathcal{N}(0,\sigma^2) \;\;\;\; i = 1,...,2N
\end{aligned}
\end{equation*}

Low values of $N$ provide a low-pass filter to seasonality effect, the proper number can be seected automatically with AIC model selection procedure. $\sigma$ is an hyper-paramter to model the flexibility of seasonality.

### Multiple seasonality trends
If the phenomenon presents differents types of periodicities (ex: weekly, monthly, yearly) I should add a different seasonality term for each periodicity I want to model with its own period.

If I have Q different overlpping periodical behaviours each one with its own period $\{ P_1, ..., P_Q\}$, I model the seasonality term as the sum of Q independent seasonality terms:

\begin{equation*}
\begin{aligned}
    s(t) & = \sum_{q=1}^{Q} s_q(t) \\
    s_q(t) & = \sum_{n=1}^{N} \left( a_n^q cos \left(\frac{2 \pi n t}{P_q} \right) + b_n^q sin \left(\frac{2 \pi n t}{P_q} \right) \right) \\
    prior: \;\; \underline{\beta}_i^q & \overset{\text{iid}}{\sim} \mathcal{N}(0,\sigma^2_q) \;\;\;\; i = 1,...,2N
\end{aligned}
\end{equation*}

$\newline$
This code implements just one seasonality term, so the observed data should reflect such periodicity. The code prophet_two_seasonalities.ipynb models two different periodicities in the same series.

### Holiday Trend
To model the impact of extraordinary events like holidays during the year, we need to provide a list of the times of these events $\{t_{D_1}, ..., t_{D_L} \}$, assigning to each one a corresponding change to the forecast $\underline{\kappa} = \{\kappa_{1}, ..., \kappa_{L} \}$.

We obtain the model:
\begin{equation*}
\begin{aligned}
    Z(t) & = \left[ \mathbb{1}(t=t_{D_1}), ..., \mathbb{1}(t=t_{D_L}) \right] \\
    h(t) & = Z(t) \underline{\kappa} \\
    prior: \;\; \kappa_i & \overset{\text{iid}}{\sim} \mathcal{N}(0,\nu^2) \;\;\;\; i = 1,...,L    
\end{aligned}
\end{equation*}

$\newline$
The days surrounding the holidays are treated as holidays themselves and included in the list of the holydays. $\tau$ is a smoothing hyper-parameetr to be set properly.

## Stan code for prediction
In prophet_two_seasonalities.ipynb eevrything is commented better.

```python
my_pred_prophet = """
functions {

  // function to obtain change point matrix for all times 
  
  matrix get_changepoint_matrix(vector t, vector t_change, int T, int S) {
    // Assumes t and t_change are sorted.
    matrix[T, S] A;
    row_vector[S] a_row;
    int cp_idx;

    // Start with an empty matrix.
    A = rep_matrix(0, T, S);
    a_row = rep_row_vector(0, S);
    cp_idx = 1;

    // Fill in each row of A.
    for (i in 1:T) {
      while ((cp_idx <= S) && (t[i] >= t_change[cp_idx])) {
        a_row[cp_idx] = 1;
        cp_idx = cp_idx + 1;
      }
      A[i] = a_row;
    }
    return A;
  }

  // Logistic trend functions

  // function to compute the adjustment factors gamma
  
  vector logistic_gamma(real k, real m, vector delta, vector t_change, int S) {
    vector[S] gamma;     // vector for adjusted offsets, for piecewise continuity
    vector[S + 1] k_s;   // vector for actual rate in each segment
    real m_pr;

    // Compute the rate in each segment
    k_s = append_row(k, k + cumulative_sum(delta));

    // Piecewise offsets (adjustment for countinuity of growth trend)
    m_pr = m; // The offset in the previous segment
    for (i in 1:S) {
      gamma[i] = (t_change[i] - m_pr) * (1 - k_s[i] / k_s[i + 1]);
      m_pr = m_pr + gamma[i];  // update for the next segment
    }
    return gamma;
  }

  // function to compute the logistic trend
  
  vector logistic_trend(
    real k,
    real m,
    vector delta,
    vector t,
    vector cap,
    matrix A,
    vector t_change,
    int S
  ) {
    vector[S] gamma;

    gamma = logistic_gamma(k, m, delta, t_change, S);
    return cap .* inv_logit((k + A * delta) .* (t - (m + A * gamma)));
  }

  // Linear trend function

  vector linear_trend(
    real k,
    real m,
    vector delta,
    vector t,
    matrix A,
    vector t_change
  ) {
    return (k + A * delta) .* t + (m + A * (-t_change .* delta));
  }

  // Flat trend function

  vector flat_trend(
    real m,
    int T
  ) {
    return rep_vector(m, T);
  }
}

/////////////////////////////////////////////////////////////////////////////////////////////
// DATA COLLECTED BY THE MODEL
data {
  int T;                // Number of time periods of the time series
  int<lower=1> K;       // Number of seasonality regressors (dimension of vector X(t) K = 2N)
  vector[T] t;          // Vector of Times of the time series
  vector[T] cap;        // Capacities for logistic trend evolving in time
  vector[T] y;          // Time series values
  int S;                // Number of changepoints
  vector[S] t_change;   // Vector of Times of trend changepoints
  matrix[T,K] X;        // Seasonality Regressor vector for each time t
  vector[K] sigmas;     // Hyper-parameter of Scale on seasonality prior
  real<lower=0> tau;    // Hyper-parameter of Scale on changepoints prior
  int trend_indicator;  // 0 for linear, 1 for logistic, 2 for flat growth trend
  vector[K] s_a;        // Indicator of additive features
  vector[K] s_m;        // Indicator of multiplicative features
  int<lower=1> T_pred;
  vector[T_pred] t_pred;
  vector[T_pred] cap_pred; 
  int S_pred;
  vector[S_pred] t_change_pred; 
  matrix[T_pred,K] X_pred;
}

transformed data {
  matrix[T, S] A;
  matrix[T_pred, S] A_pred;
  A = get_changepoint_matrix(t, t_change, T, S);    // to obtain the changepoint matrix a(t) for all t
  A_pred = get_changepoint_matrix(t_pred, t_change_pred, T_pred, S_pred);
}

////////////////////////////////////////////////////////////////////////////////////////////////////////
// PARAMETERS TO BE ESTIMATED BY THE MODEL
parameters {
  real k;                   // Base trend growth rate
  real m;                   // Trend offset
  vector[S] delta;          // Trend rate adjustments
  real<lower=0> sigma_obs;  // Observation noise
  vector[K] beta;           // Seasonality Regressor coefficients
}

transformed parameters {
  vector[T] trend;
  if (trend_indicator == 0) {
    trend = linear_trend(k, m, delta, t, A, t_change);
  } else if (trend_indicator == 1) {
    trend = logistic_trend(k, m, delta, t, cap, A, t_change, S);
  } else if (trend_indicator == 2) {
    trend = flat_trend(m, T);
  }
}

/////////////////////////////////////////////////////////////////////////////////////////////////////////
// PRIORS AND LIKELIHOOD
model {
  //priors
  k ~ normal(0, 5);                      // prior for Base trend growth rate
  m ~ normal(0, 5);                      // prior Trend offset
  delta ~ double_exponential(0, tau);    // prior Trend rate adjustments (Laplace is double exponential)
  sigma_obs ~ normal(0, 0.5);            // prior for Observation noise
  beta ~ normal(0, sigmas);              // prior for Seasonality Regressor coefficients

  // Likelihood
  y ~ normal(
  trend .* (1 + X * (beta .* s_m)) + X * (beta .* s_a),
  sigma_obs
  );
}


/////////////////////////////////////////////////////////////////////////////////////////////////
// GENERATED QUANTITIES
generated quantities {                       // vector containing the fitted values in the time range of
        vector[T+T_pred] y_pred;           // the time series + space for following next 50 predictions             
        y_pred[1:T] = y;                     // collect the observed data in the time range of the series
        vector[T_pred] trend_pred;
        if (trend_indicator == 0) {
            trend_pred = linear_trend(k, m, delta, t_pred, A_pred, t_change_pred);
        } else if (trend_indicator == 1) {
            trend_pred = logistic_trend(k, m, delta, t_pred, cap_pred, A_pred, t_change_pred, S_pred);
        } else if (trend_indicator == 2) {
            trend_pred = flat_trend(m, T_pred);
        }
        y_pred[(T+1):(T+T_pred)] = trend_pred.* (1 + X_pred * (beta .* s_m)) + X_pred * (beta .* s_a);
}

"""


stan_file = "./my_pred_prophet.stan"

with open(stan_file, "w") as f:
    print(my_pred_prophet, file=f)

my_pred_prophet = CmdStanModel(stan_file=stan_file)
```

```python
N = 50
times = np.arange(T)
cap = 1600*np.ones(T)
S = 1
S_vect = np.array([500])
P = 7
season_sigma = 10
tau = 1

T_pred = 50
times_pred = np.arange(T,T+T_pred,1)
S_pred = 1
S_vect_pred = np.array([1500])
cap_pred = 1600*np.ones(T_pred)


def season_regressor(N, times, P):
    T = times.shape[0]
    X = np.zeros((T,2*N))
    for i in np.arange(1,2*N+1,2):
        X[:,(i-1)] = np.sin(2*np.pi*times*i/P)
        X[:,i] = np.cos(2*np.pi*times*i/P)
    return X

X_train = season_regressor(N, times, P)
X_pred = season_regressor(N, times_pred, P)

pred_prophet_data = {
    "T": T,
    "K": 2*N,
    't': times,
    'cap': cap,
    'y': data,
    'S': S,
    't_change': S_vect,
    'X': X_train,
    'sigmas': season_sigma*np.ones(2*N),
    'tau': tau,
    'trend_indicator': 2,
    's_a': 1*np.ones(2*N, dtype=np.int),
    's_m': np.zeros(2*N, dtype=np.int),
    'T_pred': T_pred,
    't_pred': times_pred,
    'cap_pred': cap_pred,
    'S_pred': S_pred,
    't_change_pred': S_vect_pred,
    'X_pred': X_pred
}

N_iter = 1000
pred_prophet_fit = my_pred_prophet.sample(data=pred_prophet_data, chains=4, parallel_chains=4, seed=stan_seed,
    iter_warmup=1000, iter_sampling=N_iter, thin = 1, show_progress=True)

prophet_fit_data = az.from_cmdstanpy(pred_prophet_fit)
```

```python
az.summary(prophet_fit_data)
```

```python
az.plot_trace(prophet_fit_data, 
              var_names=["k", "m", "delta", "sigma_obs", "beta"], 
              compact=False)
plt.tight_layout()
plt.show()
```

```python
preds = prophet_fit_data.posterior.y_pred.values.reshape((4 * N_iter, len(data)+T_pred))
print(preds.shape) # for each time t, I have a vector of n_chains*N_iter of y(t)

preds_std = np.std(preds, axis = 0) # standard deviation of the prediction for each time t
preds_mean = np.mean(preds, axis = 0)
preds_std = np.column_stack((preds_mean-2*preds_std, preds_mean+2*preds_std))

fig, ax = plt.subplots(1,1)
ax.plot(data)
tnew = len(data) + np.arange(T_pred)
ax.plot(tnew, preds[:, -T_pred:].T, color="orange", alpha=0.05)
ax.plot(tnew, preds_mean[-T_pred:], color='blue', linestyle='-')
#ax.plot(tnew, preds_std[-T_pred:, :], color='red', linestyle='--')  ## plot mean +- std of prediction of y(t)
ax.set_title('Predicted values, prediction mean $\pm$ 2 std')
plt.tight_layout()
plt.show()
```
